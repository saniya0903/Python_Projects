**Import dataset**

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Loan_dataset.csv')
```

```python
df
```

```python
## checking colums
df.columns
```

```python
df.describe()
```

**Checking Null Values**

```python
df.isna().sum()
```

```python
df.isna().sum().plot(kind='bar')
```

 **Filling Null Values**

```python
for i in df.columns:
  if df[i].dtypes==object:
    df[i]=df[i].fillna(df[i].mode()[0]) ##for object values
  else:
    df[i]=df[i].fillna(df[i].mean())  ##for numerical values
```

```python
df.isna().sum()
```

**Descriptive Statistics**

```python
for i in df.columns:
    if df[i].dtypes!=object:
        print(i)
        print()
        print(df[i].describe())
        print('**************************************************************************')
```

```python
import seaborn as sns
```

```python
from matplotlib.pyplot import show
for i in df.columns:
  if df[i].dtype!=object:
    sns.histplot(data=df,x=i,kde=True)
    print(i)
    show()
```

## Correlation

```python
cat=[]
con=[]
for i in df.columns:
  if df[i].dtypes==object:
    cat.append(i)
  else:
    con.append(i)
print('cat:',cat)
print('con:',con)
```

```python
xcat=df[cat]
xcon=df[con]
```

```python
xcon.corr()     ## checking correlations
```

```python
sns.heatmap(xcon.corr(),annot=True)
```

```python
xcon.corr()['Current Credit Balance'].sort_values(ascending=False).plot(kind='bar')
```

## Pair Plot

```python
sns.pairplot(data=xcon)
```

 Advance Analysis

```python
## 1.What is the average current loan amount for each loan status?
averages=df.groupby(by='Loan Status')['Current Loan Amount'].mean()
print('Average of current loan amount for each loan status :')
print(averages)
```

```python
averages.plot(kind='bar')
plt.title('Average of current loan amount for each loan status')
plt.xlabel('Loan Status')
plt.ylabel('Average Current Loan Amount')
plt.show()
```

```python
## 2.How does the credit score vary with the annual income?
sns.scatterplot(data=df,x='Annual Income',y='Credit Score',c='purple')
plt.title('Credit Score vs Annual Income')
plt.xlabel('Annual Income')
plt.ylabel('Credit Score')
plt.show()
```

```python
## 3.Is there a correlation between the number of open accounts and the current credit balance?
correlation=df['Number of Open Accounts'].corr(df['Current Credit Balance'])
print('The correlation between the number of open accounts and the current credit balance is:',)
print(correlation)
```

```python
## 0 indicates no linear relationship
```
